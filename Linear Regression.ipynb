```
### File Location and Type
file_location = "/FileStore/tables/tips.csv"
file_type = "csv"

### Loading the dataset into notebook
df = spark.read.csv(file_location, header=True, inferSchema=True)
```

```
df.show()
```

```
df.printSchema()
```

```
df.columns
```

```
### Handling The categorical data
from pyspark.ml.feature import StringIndexer
```

```
indexer = StringIndexer(inputCols=["sex", "smoker", "day", "time"], outputCols=["sex_indexed", "smoker_indexed", "day_indexed", "time_indexed"])

df_new = indexer.fit(df).transform(df)
df_new.show()
```

```
from pyspark.ml.feature import VectorAssembler

feature_assembler = VectorAssembler(inputCols=["tip", "size", "sex_indexed", "smoker_indexed", "day_indexed", "time_indexed"], outputCol="IndependentFeatures")

output = feature_assembler.transform(df_new)
```

```
output.select("IndependentFeatures").show()
```

```
output.show()
```

```
### Creating final data

finalized_data = output.select("IndependentFeatures", "total_bill")
```

```
finalized_data.show()
```

```
### Implementing The Machine Learning Model Now Since The Data is Prepered To Fed To Machine

from pyspark.ml.regression import LinearRegression

### Train Test Split
train_data, test_data = finalized_data.randomSplit([0.80, 0.20])
regressor = LinearRegression(featuresCol="IndependentFeatures", labelCol="total_bill")
regressor = regressor.fit(train_data)
```

```
regressor.coefficients
```

```
regressor.intercept
```

```
### Predictions
pred_results=regressor.evaluate(test_data)
```

```
## Final comparison
pred_results.predictions.show()
```

```
### Performance Metrics
pred_results.r2,pred_results.meanAbsoluteError,pred_results.meanSquaredError
```
